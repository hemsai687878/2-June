```python
import pandas as pd
```

```python
df = pd.read_csv("cust_satisfaction.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df
```

```python
df['Customer Type'].unique()
```

```python
Cust_dict={'Loyal Customer':1, 'disloyal Customer':2}
```

```python
Cust_dict
```

```python
df['Customer Type'] = df['Customer Type'].map(Cust_dict)
```

```python
df['Customer Type']
```

```python
df['Type of Travel'].unique()
```

```python
travel_dict={'Personal Travel':1, 'Business travel':2}
```

```python
travel_dict
```

```python
df['Type of Travel'] = df['Type of Travel'].map(travel_dict)
```

```python
df['Type of Travel']
```

```python
df['Gender'].unique()
```

```python
gender_dict = {'Male':1, 'Female':2}
```

```python
df['Gender'] = df['Gender'].map(gender_dict)
```

```python
df['Gender']
```

```python
df['Class'].unique()
```

```python
class_dict={'Eco':1,'Eco Plus':2,'Business':3}
```

```python
df['Class'] = df['Class'].map(class_dict)
```

```python
df['Class']
```

```python
df
```

```python
df['satisfaction'].unique()
```

```python
satis_dict={'neutral or dissatisfied':1, 'satisfied':2}
```

```python
df['satisfaction']= df['satisfaction'].map(satis_dict)
```

```python
df
```

```python
X = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression

from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python

lin_pred=lin_reg.predict(X_test)
```

```python
lin_pred
```

```python

print("training score:- ",lin_reg.score(X_train, y_train))
```

```python

print("testing score:- ",lin_reg.score(X_test, y_test))
```

```python
lin_mse = mean_squared_error(y_test, lin_pred)
```

```python
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)
```

```python
log_pred=log_reg.predict(X_test)
```

```python
log_pred
```

```python

print("training score:- ",log_reg.score(X_train, y_train))
```

```python

print("testing score:- ",log_reg.score(X_test, y_test))
```

```python

log_acc = accuracy_score(y_test, log_pred)
```

```python
print("Linear Regression MSE:", lin_mse)
print("Logistic Regression Accuracy:", log_acc)
```
